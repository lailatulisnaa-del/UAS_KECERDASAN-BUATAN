```python
import random
import matplotlib.pyplot as plt

# ==========================================
# 1. DATA INPUT (Poin 2 Soal)
# ==========================================

courses = [
    ("AI", "DosenA"), ("ML", "DosenB"), ("DB", "DosenC"),
    ("CN", "DosenD"), ("SE", "DosenE"), ("OS", "DosenF"),
    ("IR", "DosenA"), ("NLP", "DosenB"), ("CV", "DosenC"),
    ("IOT", "DosenD"), ("HCI", "DosenE"), ("CyberSec", "DosenF"),
    ("BigData", "DosenA"), ("DL", "DosenB"), ("NoSQL", "DosenC"),
    ("Cloud", "DosenD"), ("UX", "DosenE"), ("SecEng", "DosenF"),
    ("Agile", "DosenA"), ("RL", "DosenB"), ("ETL", "DosenC"),
    ("DevOps", "DosenD"), ("GameDev", "DosenE"), ("Blockchain", "DosenF")
]

rooms = ["R1", "R2", "R3", "R4"]

timeslots = [
    "Senin-08", "Senin-10", "Senin-13",
    "Selasa-08", "Selasa-10", "Selasa-13",
    "Rabu-08", "Rabu-10", "Rabu-13",
    "Kamis-08", "Kamis-10", "Kamis-13"
]

# ==========================================
# 2. PARAMETER ALGORITMA GENETIKA (Poin 4)
# ==========================================

POPULATION_SIZE = 60
GENERATIONS = 100
CROSSOVER_RATE = 0.85
MUTATION_RATE = 0.20

# ==========================================
# 3. EVALUASI KONFLIK & FITNESS (Poin 5)
# ==========================================

def count_conflicts(individual):
    conflicts = 0
    room_time_map = {}
    lecturer_time_map = {}
    lecturer_day_map = {}

    for i, (course_name, lecturer) in enumerate(courses):
        room_idx, time_idx = individual[i]
        room = rooms[room_idx]
        timeslot = timeslots[time_idx]
        day = timeslot.split("-")[0]

        # 1. Konflik Ruang
        rt_key = (room, timeslot)
        if rt_key in room_time_map:
            conflicts += 1
        else:
            room_time_map[rt_key] = course_name

        # 2. Konflik Dosen-Waktu
        lt_key = (lecturer, timeslot)
        if lt_key in lecturer_time_map:
            conflicts += 1
        else:
            lecturer_time_map[lt_key] = course_name

        # 3. Konflik Dosen-Hari
        ld_key = (lecturer, day)
        if ld_key in lecturer_day_map:
            conflicts += 1
        else:
            lecturer_day_map[ld_key] = course_name

    return conflicts

def calculate_fitness(individual):
    conflicts = count_conflicts(individual)
    fitness = 100 / (1 + conflicts)
    return fitness, conflicts

# ==========================================
# 4. OPERATOR GENETIKA (Poin 3 & 4)
# ==========================================

def create_individual():
    return [
        (random.randint(0, len(rooms) - 1), random.randint(0, len(timeslots) - 1))
        for _ in range(len(courses))
    ]

def crossover(parent1, parent2):
    if random.random() < CROSSOVER_RATE:
        point = random.randint(1, len(courses) - 1)
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
        return child1, child2
    return parent1.copy(), parent2.copy()

def mutate(individual):
    mutated = []
    for gene in individual:
        if random.random() < MUTATION_RATE:
            new_room = random.randint(0, len(rooms) - 1)
            new_time = random.randint(0, len(timeslots) - 1)
            mutated.append((new_room, new_time))
        else:
            mutated.append(gene)
    return mutated

def select_parent(population, fitnesses):
    tournament_size = 3
    selected_indices = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected_indices, key=lambda idx: fitnesses[idx])
    return population[best_idx]

# ==========================================
# 5. PROSES EVOLUSI & CETAK PER GENERASI (Poin 8)
# ==========================================

population = [create_individual() for _ in range(POPULATION_SIZE)]

max_fitness_history = []
avg_fitness_history = []

for gen in range(1, GENERATIONS + 1):
    evaluations = [calculate_fitness(ind) for ind in population]
    fitnesses = [eval_res[0] for eval_res in evaluations]

    max_fit = max(fitnesses)
    avg_fit = sum(fitnesses) / POPULATION_SIZE

    max_fitness_history.append(max_fit)
    avg_fitness_history.append(avg_fit)

    # Cetak perkembangan per generasi (Poin 8)
    print(f"Gen {gen:4d} : Max Fitness = {max_fit:6.2f} | Avg Fitness = {avg_fit:6.2f}")

    new_population = []
    best_idx = fitnesses.index(max_fit)
    new_population.append(population[best_idx])  # Elitisme

    while len(new_population) < POPULATION_SIZE:
        parent1 = select_parent(population, fitnesses)
        parent2 = select_parent(population, fitnesses)
        child1, child2 = crossover(parent1, parent2)
        new_population.append(mutate(child1))
        if len(new_population) < POPULATION_SIZE:
            new_population.append(mutate(child2))

    population = new_population

# Evaluasi generasi akhir
evaluations = [calculate_fitness(ind) for ind in population]
fitnesses = [eval_res[0] for eval_res in evaluations]
best_idx = fitnesses.index(max(fitnesses))

best_schedule = population[best_idx]
best_fitness, best_conflicts = evaluations[best_idx]

# ==========================================
# 6. HASIL AKHIR & JADWAL TERBAIK (Poin 6 & 7)
# ==========================================

print("\n=== JADWAL TERBAIK ===")
for i, (course_name, lecturer) in enumerate(courses):
    room_idx, time_idx = best_schedule[i]
    room = rooms[room_idx]
    timeslot = timeslots[time_idx]
    print(f"MK: {course_name:<10} | Ruang: {room:<2} | Waktu: {timeslot:<9} | Dosen: {lecturer}")

print("\n=== HASIL EVALUASI AKHIR ===")
print(f"Fitness Terbaik : {best_fitness:.2f}")
print(f"Jumlah Konflik  : {best_conflicts}")

# ==========================================
# 7. VISUALISASI GRAFIK MATPLOTLIB (Poin 9)
# ==========================================

plt.figure(figsize=(10, 5))

# Plot Fitness Maksimum & Rata-rata
plt.plot(range(1, GENERATIONS + 1), max_fitness_history, label="Fitness Maksimum")
plt.plot(range(1, GENERATIONS + 1), avg_fitness_history, label="Fitness Rata-rata")

# Pengaturan Judul & Label Persis Soal Poin 9
plt.title("Perkembangan Fitness - Algoritma Genetika Penjadwalan Mata Kuliah")
plt.xlabel("Generasi")
plt.ylabel("Fitness")
plt.legend(loc="upper left")
plt.grid(True)

plt.tight_layout()
plt.show()
```
